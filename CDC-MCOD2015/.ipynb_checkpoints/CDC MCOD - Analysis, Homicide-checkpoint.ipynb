&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Import the data:

```python
X = pd.read_csv('https://s3.amazonaws.com/cdc-mortality/mort_featureset.csv')

len(X)
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Set up the slice of original dataframe we're interested in:

```python
boolean_slicing = (X['homicide'] == 1)

dropped_columns = [u'accident', u'homicide', u'natural', u'suicide']

filename = 'Zhom'

title = 'Homicide'

max_clusters = 10
```

```python
X_new = X[boolean_slicing]

X_new.drop(dropped_columns, axis = 1, inplace = True)
```

```python
len(X_new)
```

### Hierarchical Clustering

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;First, we need to calculate the linkages.  (See written hierarchical clustering tutorial [here](https://joernhees.de/blog/2015/08/26/scipy-hierarchical-clustering-and-dendrogram-tutorial/), video tutorial [here](https://www.youtube.com/watch?v=EQZaSuK-PHs).)

```python
start = datetime.datetime.now()
print start

Z = linkage(X_new, 'ward')

print datetime.datetime.now() - start
```

```python
len(Z)
```

```python
pd.DataFrame(Z).to_csv('./assets/{}.csv'.format(filename), index = False)
```

```python
Z = np.array(pd.read_csv('./assets/{}.csv'.format(filename)))
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;See explanation of Cophenetic distances [here](https://en.wikipedia.org/wiki/Cophenetic_correlation).  Closer to 1 is better.

```python
start = datetime.datetime.now()
print start 

c, _ = cophenet(Z, pdist(X_new))
print datetime.datetime.now() - start
c
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The first linkage in this dendrogram occurs between observations indexed as the first two values; these points are a distance the third value away from each other and comprise the fourth value number of points when clustered:

```python
Z[1]
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Likewise, for the final linkage:

```python
Z[-1]
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Behold, a truncated dendrogram!  Labels are either size of cluster or index of a single observation.

```python
plt.figure(figsize=(25, 10))
plt.title('CDC MCOD: Truncated Hierarchical Clustering Dendrogram, {}'.format(title), fontsize = 30, y = 1.03)
plt.xlabel('sample index or (cluster size)', fontsize = 24)
plt.ylabel('distance', fontsize = 24)
plt.yticks(fontsize = 18)
dendrogram(
    Z,
    truncate_mode = 'lastp',  # only show cluster merges after there are p clusters
    p = max_clusters,  # choose p
    show_leaf_counts = True,  
    leaf_rotation = 0,
    leaf_font_size = 20.,
    show_contracted = True,  # shows marks where previous merges would've occured if not for truncation
)
plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Use [elbow method](https://joernhees.de/blog/2015/08/26/scipy-hierarchical-clustering-and-dendrogram-tutorial/) -- with caution! -- to suggest number of clusters:

```python
plt.figure(figsize = (12, 8))

last = Z[-20:, 2]
last_rev = last[::-1]
idxs = np.arange(1, len(last) + 1)
plt.plot(idxs, last_rev)

acceleration = np.diff(last, 2)  # 2nd derivative of the distances
acceleration_rev = acceleration[::-1]
plt.plot(idxs[:-2] + 1, acceleration_rev)

plt.yticks(fontsize = 13)
plt.xticks(range(0, 24, 2), fontsize = 13)
plt.show()
k = acceleration_rev.argmax() + 2  # if idx 0 is the max of this we want 2 clusters
print "clusters:", k
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Get cluster labels using desired number of clusters:**

```python
k_clusters = fcluster(Z, k, criterion='maxclust')
k_clusters
```

```python
plt.style.use('fivethirtyeight')
fig, ax1 = plt.subplots(figsize=(12, 8))

data = []
for i in range(1, k + 1):
    data.append(MinMaxScaler([1, 100]).fit_transform(X_new['age'].reshape(-1,1))[k_clusters == i])

plt.boxplot(data, 0, 'gD')

ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)

ax1.set_axisbelow(True)
ax1.set_title('CDC MCOD: Age at Time of Death, {} Clusters'.format(title), y = 1.03, fontsize = 24)
ax1.set_xlabel('Cluster Number', fontsize = 20)
ax1.set_ylabel('Age (years)', fontsize = 20)

# Set the axes ranges and axes labels
numBoxes = k
ax1.set_xlim(0.5, numBoxes + 0.5)
ax1.set_ylim(0, 105)
xtickNames = plt.setp(ax1, xticklabels = range(1, k + 1))
plt.setp(xtickNames, fontsize=20)
ytickNames = plt.setp(ax1, yticklabels = range(0, 120, 20))
plt.setp(ytickNames, fontsize=14)

plt.axhline(0, color='black', alpha = 0.5)

del data

plt.show()
```

### PCA

```python
pca = PCA().fit(X_new)
```

```python
pca_df = pd.DataFrame(zip(pca.components_, pca.explained_variance_/sum(pca.explained_variance_)), \
                      columns = ['eigenvector', 'eigenvalue'])

pca_df.head()
```

```python
plt.figure(figsize=(12,8))
plt.style.use('fivethirtyeight')

pca_df['eigenvalue'].plot.bar(color = 'slateblue', alpha = 0.6)

plt.title('CDC MCOD: Variance Explained, {}'.format(title), fontsize = 24)

plt.ylabel('Variance Explained', fontsize = 18)
#plt.yticks(range(0, 9), fontsize = 13)
plt.ylim([0, 1.2*pca_df['eigenvalue'].max()])

plt.xlabel('Principal Component #', fontsize = 18)
plt.axhline(0, color='black', alpha = 0.6)
plt.xticks([])

plt.grid(which = 'major', axis = 'x')

plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;How much variance is explained using the first three principal components?

```python
pca_df.loc[0:3, 'eigenvalue'].sum()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,8))

cum_var_exp = np.cumsum(pca.explained_variance_ratio_)*100

plt.plot(range(0, len(X_new.columns)), cum_var_exp, linewidth = 3, color = 'navy', alpha = 0.6)
plt.axhline(y=0, linewidth=2, color='grey', ls='solid')
plt.axhline(y=100, linewidth=1, color='k', ls='dashed')
plt.xlim([-1,len(X.columns)])
plt.xticks(fontsize = 13)
plt.ylim([-5,105])
plt.yticks(range(0, 120, 20), fontsize = 13)
plt.ylabel('Cumulative Variance Explained', fontsize=16)
plt.xlabel('Principle Component #', fontsize=16)  
plt.title('CDC MCOD: Cumulative Variance Explained, {}'.format(title), fontsize=20, y = 1.03)
plt.show()
```

```python
X_pca = pd.DataFrame(np.dot(pca.components_[0:3], X_new.T)).T
X_pca.columns = ['PC1', 'PC2', 'PC3']

X_pca['labels'] = k_clusters
X_pca.head(2)
```

```python
fig = pylab.figure(figsize = (12, 8))
ax = Axes3D(fig)

ax.scatter(X_pca['PC1'], X_pca['PC2'], X_pca['PC3'], zdir='z', alpha=.7, c = k_clusters, cmap = 'jet')

ax.set_title("CDC MCOD: {} Clusters (PCA)".format(title), fontsize = 24, y = 1.03)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Inspect the clusters!

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Construct a dataframe with the clusters as columns, rows as features, and cell entries as average feature value from the original dataframe (column sums, divided by number of rows):

```python
big_df = pd.DataFrame()
for i in range(1, k + 1):
    lil_cluster = X_new[k_clusters == i]
    length = len(lil_cluster)
    lil_df = pd.DataFrame(lil_cluster.sum()/len(lil_cluster), columns = ['cluster {}'.format(i)])
    big_df = pd.concat([big_df, lil_df], axis = 1)
```

```python
plt.figure(figsize = (12, 20))
sns.heatmap(big_df, cmap = 'coolwarm')
plt.yticks(fontsize = 12)
plt.xticks(fontsize = 18)
plt.show()
```

```python
print 'AVG AGE'
print 'cluster|mean|stdev'
for i in range(1, k + 1):
    print i, round(pd.DataFrame(ages[k_clusters == i])[0].mean(), 0), \
    round(pd.DataFrame(ages[k_clusters == i])[0].std())
```

```python
plt.figure(figsize = (12, 8))

for i in range(1, kk + 1):
    lil_ages = ages[k_clusters == i]
    plt.hist(lil_ages, alpha = 0.3, label = 'cluster {}'.format(i), color = cm.Set1(i/float(k)))
    
    density = gaussian_kde(lil_ages)
    xs = np.linspace(lil_ages.min(), lil_ages.max(), 1000)
    density.covariance_factor = lambda : 0.25
    density._compute_covariance()
    plt.plot(xs, density(xs)*len(lil_ages)*8, linewidth = 2.5, alpha = 0.85, color = cm.Set1(i/float(k)))

plt.title('CDC MCOD: Histogram of Age, {} Clusters'.format(title), fontsize = 24, y = 1.03)
plt.xticks(fontsize = 14)
plt.xlabel('Age (years)', fontsize = 18)
plt.yticks(fontsize = 14)
plt.ylabel('Count', fontsize = 18)
    
lgnd = plt.legend(loc = 'best', fontsize = 14)

for lil in range(0, k):
    lgnd.legendHandles[lil]._alpha = 0.8

plt.grid(which = 'major', axis = 'x')

plt.show()
```

```python
for i in range(1, kk + 1):
    clustoid = X_new[k_clusters == i]

    print 'Cluster {} makes up {} of {} dataframe.'.format(i, round(len(clustoid)/float(len(X_new)), 2), title)

    plt.figure(figsize = (12, 8))
    corr_mat = np.tril(clustoid.corr(), k=-1)
    sns.heatmap(corr_mat, cmap = 'seismic')
    plt.title('CDC MCOD: Feature Correlations, {} Cluster {}'.format(title, i), fontsize = 20, y = 1.03)
    plt.xticks(np.arange(0.5, len(clustoid.columns) + 0.5, 1), clustoid.columns, fontsize = 7.5, rotation = 90)
    plt.yticks(np.arange(0.5, len(clustoid.columns) + 0.5, 1), list(clustoid.columns)[::-1], fontsize = 7.5, rotation = 0)
    plt.show()
```
