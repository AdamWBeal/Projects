&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;See explanation of Cophenetic distances [here](https://en.wikipedia.org/wiki/Cophenetic_correlation).  Closer to 1 is better.

```python
start = datetime.datetime.now()
print start, '\n' 

Z_dict = {
    'ward' : Zwar,
    'single' : Zsin, 
    'complete' : Zcom, 
    'average' : Zavg, 
    'weighted' : Zwtd, 
    'centroid' : Zcen, 
    'median' : Zmed
}

Z_list = []
method = []
for i in Z_dict.keys():
    c, d = cophenet(Z_dict[i], pdist(codes_final))  ## c=cophenetic corr distance; d=cophenetic distance matrix
    Z_list.append(round(c, 4))
    method.append(i)
    print i, datetime.datetime.now() - start
    
Z_df = pd.DataFrame(Z_list, columns = ['c'], index = method).sort_values(by = 'c', ascending = False)
```

```python
plt.figure(figsize = (12,8))

Z_df['c'].plot.bar()

plt.title('Cophenetic Distance for Various Linkage Methods', fontsize = 24, y = 1.03)

plt.xlabel('Linkage Method', fontsize = 18)
plt.xticks(fontsize = 14, rotation = 0)

plt.ylim([0, 1])
plt.ylabel('Score', fontsize = 18)
plt.yticks(np.arange(0, 1.1, 0.1), fontsize = 14)
plt.grid(axis = 'x', which = 'major')

plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Use [elbow method](https://joernhees.de/blog/2015/08/26/scipy-hierarchical-clustering-and-dendrogram-tutorial/) -- with caution! -- to suggest number of clusters:

```python
plt.figure(figsize = (12, 8))

last = Z[-20:, 2]
last_rev = last[::-1]
idxs = np.arange(1, len(last) + 1)
plt.plot(idxs, last_rev)

acceleration = np.diff(last, 2)  # 2nd derivative of the distances
acceleration_rev = acceleration[::-1]
plt.plot(idxs[:-2] + 1, acceleration_rev)

plt.yticks(fontsize = 13)
plt.xticks(range(0, 24, 2), fontsize = 13)
plt.show()
kk = acceleration_rev.argmax() + 2  # if idx 0 is the max of this we want 2 clusters
print "clusters:", kk
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Get cluster labels using desired number of clusters:**

```python
k = kk
k_clusters = fcluster(Z, k, criterion='maxclust')
k_clusters
```

### PCA

```python
pca = PCA().fit(X_big)
```

```python
pca_df = pd.DataFrame(zip(pca.components_, pca.explained_variance_/sum(pca.explained_variance_)), \
                      columns = ['eigenvector', 'eigenvalue'])

pca_df.head()
```

```python
plt.figure(figsize=(12,8))
plt.style.use('fivethirtyeight')

pca_df['eigenvalue'].plot.bar(color = 'slateblue', alpha = 0.6)

plt.title('CDC MCOD: Variance Explained, {}'.format(title), fontsize = 24)

plt.ylabel('Variance Explained', fontsize = 18)
#plt.yticks(range(0, 9), fontsize = 13)
plt.ylim([0, 1.2*pca_df['eigenvalue'].max()])

plt.xlabel('Principal Component #', fontsize = 18)
plt.axhline(0, color='black', alpha = 0.6)
plt.xticks([])

plt.grid(which = 'major', axis = 'x')

plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,8))

cum_var_exp = np.cumsum(pca.explained_variance_ratio_)*100

plt.plot(range(0, len(X_big.columns)), cum_var_exp, linewidth = 3, color = 'navy', alpha = 0.6)
plt.axhline(y=0, linewidth=2, color='grey', ls='solid')
plt.axhline(y=100, linewidth=1, color='k', ls='dashed')
plt.xlim([-10,len(X_big.columns)])
plt.xticks(fontsize = 13)
plt.ylim([-5,105])
plt.yticks(range(0, 120, 20), fontsize = 13)
plt.ylabel('Cumulative Variance Explained', fontsize=16)
plt.xlabel('Principle Component #', fontsize=16)  
plt.title('CDC MCOD: PCA Cumulative Variance Explained, {}'.format(title), fontsize=20, y = 1.03)
plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Inspect the clusters!

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Construct a dataframe with the clusters as columns, rows as features, and cell entries as average feature value from the original dataframe (column sums, divided by number of rows):

```python
big_df = pd.DataFrame()
for i in range(1, k + 1):
    lil_cluster = X_big[k_clusters == i]
    length = len(lil_cluster)
    lil_df = pd.DataFrame(lil_cluster.sum()/len(lil_cluster), columns = ['cluster {}'.format(i)])
    big_df = pd.concat([big_df, lil_df], axis = 1)
    
big_df = big_df[big_df.sum(axis = 1) != 0]
```

```python
plt.figure(figsize = (12, 20))
sns.heatmap(big_df.iloc[0:100, :], cmap = 'coolwarm')

plt.title('CDC MCOD: Clustered Features, {}'.format(title), fontsize = 24)
plt.yticks(fontsize = 8)
plt.xticks(fontsize = 18)
plt.show()
```

```python
label_dict = {
    'marstat' : 'Marital Status',
    'placdth' : 'Place of Death', 
    'monthdth' : 'Month of Death',
    'weekday' : 'Day of Week',
    'mandeath' : 'Manner of Death',
    'restat' : 'Residential Status',
    'educ2003' : 'Education Level'
}

for category in feature_dict.keys():
    try:
        if category == 'codes':
            pass
        else:
            data = pd.DataFrame(StandardScaler().fit_transform(big_df.loc[feature_dict[category], :]), 
                index = feature_dict[category], columns = big_df.columns).sort_values(by = 'cluster 1', 
                ascending = False)
            if category == 'monthdth':
                data = data.reindex(index = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 
                                             'August', 'September', 'October', 'November', 'December'])
            elif category == 'weekday':
                data = data.reindex(index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 
                                             'Sunday'])
            elif category == 'educ2003':
                data = data.reindex(index = [u'8th grade or less', 
                                            u'9 - 12th grade, no diploma',
                                            u'high school graduate or GED completed', 
                                            u'some college credit, but no degree',
                                            u'Associate degree', 
                                            u'Bachelors degree', 
                                            u'Masters degree', 
                                            u'Doctorate or professional degree'])
            plt.figure(figsize = (12, 8))            
            data.plot.bar(figsize = (10,6), colors = ['darkred', 'goldenrod', 'steelblue'], alpha = 0.6)
            plt.title('CDC MCOD: Feature Proportion, {}'.format(title))
            plt.xticks(fontsize = 14, rotation = -90)
            plt.xlabel(label_dict[category], fontsize = 18)
            plt.yticks(fontsize = 14)
            plt.ylabel('Standardized Proportion', fontsize = 18)
            plt.grid(True)
            plt.grid(which = 'major', axis = 'x')
            plt.legend(loc = 'best', fontsize = 13)
            plt.axhline(0, color = 'k', alpha = 0.6, lw = 1.5)
            plt.show()
    except: 
        pass
```

```python
print 'AVG AGE'
print 'cluster|mean|stdev'
for i in range(1, k + 1):
    print i, round(pd.DataFrame(ages[k_clusters == i])[0].mean(), 0), \
    round(pd.DataFrame(ages[k_clusters == i])[0].std())
```

```python
plt.figure(figsize = (12, 8))

for i in range(1, kk + 1):
    lil_ages = ages[k_clusters == i]
    plt.hist(lil_ages, alpha = 0.3, label = 'cluster {}'.format(i), color = cm.Set1(i/float(k)), normed = 1)
    
    ## draw the gaussian KDE
    density = gaussian_kde(lil_ages)
    xs = np.linspace(lil_ages.min(), lil_ages.max(), 1000)
    density.covariance_factor = lambda : 0.25
    density._compute_covariance()
    plt.plot(xs, density(xs), linewidth = 2.5, alpha = 0.85, color = cm.Set1(i/float(k)))

plt.title('CDC MCOD: Histogram of Age, {} ({} clusters)'.format(title, kk), fontsize = 24, y = 1.03)
plt.xticks(fontsize = 14)
plt.xlabel('Age (years)', fontsize = 18)
plt.yticks(fontsize = 14)
plt.ylabel('Normed Count (individual)', fontsize = 18)
    
lgnd = plt.legend(loc = 'best', fontsize = 14)

for lil in range(0, k):
    lgnd.legendHandles[lil]._alpha = 0.8

plt.grid(which = 'major', axis = 'x')

plt.show()
```
