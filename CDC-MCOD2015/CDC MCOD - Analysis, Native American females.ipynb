## Cluster Analysis:  Native American Females

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Data cleaning is in a separate [notebook](https://github.com/forzavitale/Projects/blob/master/CDC-Mortality2015/CDC%20MCOD%20-%20Cleaning.ipynb); cleaned set available on AWS [here]('https://s3.amazonaws.com/cdc-mortality/mort_featureset.csv').    

```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.spatial import distance
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet, fcluster
from scipy.spatial.distance import pdist
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN, KMeans
from sklearn.decomposition import PCA 
from mpl_toolkits.mplot3d import Axes3D
import pylab
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Import the data:

```python
X = pd.read_csv('https://s3.amazonaws.com/cdc-mortality/mort_featureset.csv')

len(X)
```

```python
X.head(2)
```

```python
X.columns
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Let's consider a subset of the observations, say, Native American females:**

```python
Xnaf = X[(X['amer_indian'] == 1) & (X['sex'] == 0)]

Xnaf.drop(['amer_indian', 'asian/PI', 'black', 'white', 'sex'], axis = 1, inplace = True)
```

```python
len(Xnaf)
```

### Hierarchical Clustering

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;First, we need to calculate the linkages.  (See written hierarchical clustering tutorial [here](https://joernhees.de/blog/2015/08/26/scipy-hierarchical-clustering-and-dendrogram-tutorial/), video tutorial [here](https://www.youtube.com/watch?v=EQZaSuK-PHs).)

```python
start = datetime.datetime.now()
print start

Z = linkage(Xnaf, 'ward')

print datetime.datetime.now() - start
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;See explanation of Cophenetic distances [here](https://en.wikipedia.org/wiki/Cophenetic_correlation).  Closer to 1 is better.

```python
c, _ = cophenet(Z, pdist(Xnaf))
c
```

```python
len(Z)
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The first linkage in this dendrogram occurs between observations 3378 and 5530; they are a very tiny distance away from each other and comprise two points when clustered:

```python
Z[1]
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The last linkage in this dendrogram occurs between clusters 14136 and 14141, which are approximately 69 units apart and collectively contain 7072 observations.

```python
Z[-1]
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Behold, a dendrogram!  Index numbers as leaf labels by default.  

```python
plt.figure(figsize=(25, 10))
plt.title('CDC MCOD: Hierarchical Clustering Dendrogram, Native American Women', fontsize = 30, y = 1.03)
plt.xlabel('sample index', fontsize = 24)
plt.xticks([])
plt.ylabel('distance', fontsize = 24)
plt.yticks(fontsize = 18)
dendrogram(
    Z,
    leaf_rotation=90., 
    leaf_font_size=8., 
)
plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Now, a truncated dendrogram!  Labels are either size of cluster or index of a single observation.

```python
plt.figure(figsize=(25, 10))
plt.title('CDC MCOD: Hierarchical Clustering Dendrogram (truncated), Native American Women', fontsize = 30, y = 1.03)
plt.xlabel('sample index or (cluster size)', fontsize = 24)
plt.ylabel('distance', fontsize = 24)
plt.yticks(fontsize = 18)
dendrogram(
    Z,
    truncate_mode = 'lastp',  # only show joins after there are p clusters
    p = 10,  # choose p
    show_leaf_counts = True,  
    leaf_rotation = 0,
    leaf_font_size = 20.,
    show_contracted = True,  # shows marks where previous splits would've branched (without truncation)
)
plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Use [elbow method](https://joernhees.de/blog/2015/08/26/scipy-hierarchical-clustering-and-dendrogram-tutorial/) -- with caution! -- to suggest number of clusters:

```python
plt.figure(figsize = (12, 8))

last = Z[-30:, 2]
last_rev = last[::-1]
idxs = np.arange(1, len(last) + 1)
plt.plot(idxs, last_rev)

acceleration = np.diff(last, 2)  # 2nd derivative of the distances
acceleration_rev = acceleration[::-1]
plt.plot(idxs[:-2] + 1, acceleration_rev)

plt.yticks(fontsize = 13)
plt.xticks(range(0, 32, 2), fontsize = 13)
plt.show()
k = acceleration_rev.argmax() + 2  # if idx 0 is the max of this we want 2 clusters
print "clusters:", k
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;It's suggesting six, with four close behind; let's try six.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Get cluster labels using distance:**

```python
max_d = 30 ## at distance = 30 there are six clusters in the dendrogram
dist_clusters = fcluster(Z, max_d, criterion='distance')
dist_clusters
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Get cluster labels using desired number of clusters:**

```python
k=6
k_clusters = fcluster(Z, k, criterion='maxclust')
k_clusters
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Confirm distance criterion method returned desired number of clusters:

```python
set(k_clusters - dist_clusters)
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Yep.

```python
plt.style.use('fivethirtyeight')
fig, ax1 = plt.subplots(figsize=(12, 8))

data = []
for i in range(1, 7):
    data.append(Xnaf.loc[k_clusters == i, 'age'])

plt.boxplot(data, 0, 'gD')

ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)

ax1.set_axisbelow(True)
ax1.set_title('CDC MCOD: Age at Time of Death, Native American Female Clusters', y = 1.03, fontsize = 24)
ax1.set_xlabel('Cluster', fontsize = 20)
ax1.set_ylabel('Scaled Age', fontsize = 20)

# Set the axes ranges and axes labels
numBoxes = k
ax1.set_xlim(0.5, numBoxes + 0.5)
ax1.set_ylim(0, 1)
xtickNames = plt.setp(ax1, xticklabels = range(1,7))
plt.setp(xtickNames, fontsize=20)
ytickNames = plt.setp(ax1, yticklabels = [])
plt.setp(ytickNames, fontsize=14)

plt.axhline(0, color='black', alpha = 0.5)

del data

plt.show()
```

### PCA

```python
pca = PCA().fit(Xnaf)
```

```python
pca_df = pd.DataFrame(zip(pca.components_, pca.explained_variance_/sum(pca.explained_variance_)), \
                      columns = ['eigenvector', 'eigenvalue'])

pca_df.head()
```

```python
plt.figure(figsize=(12,8))
plt.style.use('fivethirtyeight')

pca_df['eigenvalue'].plot.bar(color = 'slateblue', alpha = 0.6)

plt.title('CDC MCOD: Variance Explained, Native American Women', fontsize = 24)

plt.ylabel('Variance Explained', fontsize = 18)
#plt.yticks(range(0, 9), fontsize = 13)
plt.ylim([0, 1.2*pca_df['eigenvalue'].max()])

plt.xlabel('Principal Component #', fontsize = 18)
plt.axhline(0, color='black', alpha = 0.6)
plt.xticks([])

plt.grid(which = 'major', axis = 'x')

plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;How much variance is explained using the first three principal components?

```python
pca_df.loc[0:3, 'eigenvalue'].sum()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Only about a quarter!  We won't do anything about that, now.  Onward.

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,8))

cum_var_exp = np.cumsum(pca.explained_variance_ratio_)*100

plt.plot(range(0, len(X.columns)), cum_var_exp, linewidth = 3, color = 'navy', alpha = 0.6)
plt.axhline(y=0, linewidth=2, color='grey', ls='solid')
plt.axhline(y=100, linewidth=1, color='k', ls='dashed')
plt.xlim([-1,len(X.columns)])
plt.xticks(fontsize = 13)
plt.ylim([-5,105])
plt.yticks(range(0, 120, 20), fontsize = 13)
plt.ylabel('Cumulative Variance Explained', fontsize=16)
plt.xlabel('Principle Component #', fontsize=16)  
plt.title('CDC MCOD: Cumulative Variance Explained, Native American Women', fontsize=20, y = 1.03)
plt.show()
```

```python
X_pca = pd.DataFrame(np.dot(pca.components_[0:3], Xnaf.T)).T
X_pca.columns = ['PC1', 'PC2', 'PC3']

X_pca['labels'] = k_clusters
X_pca.head(2)
```

```python
fig = pylab.figure(figsize = (12, 8))
ax = Axes3D(fig)

ax.scatter(X_pca['PC1'], X_pca['PC2'], X_pca['PC3'], zdir='z', alpha=.7, c = k_clusters, cmap = 'jet')

ax.set_title("CDC MCOD: Native American Women Clusters (PCA)", fontsize = 24, y = 1.03)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Construct a dataframe with the clusters as columns, rows as features, and cell entries as average feature value from the original dataframe (column sums, divided by number of rows):

```python
big_df = pd.DataFrame()
for i in range(1, k + 1):
    lil_cluster = Xnaf[k_clusters == i]
    length = len(lil_cluster)
    lil_df = pd.DataFrame(lil_cluster.sum()/len(lil_cluster), columns = ['cluster{}'.format(i)])
    big_df = pd.concat([big_df, lil_df], axis = 1)
```

```python
plt.figure(figsize = (12, 20))
sns.heatmap(big_df, cmap = 'coolwarm')
plt.yticks(fontsize = 12)
plt.xticks(fontsize = 12)
plt.show()
```
